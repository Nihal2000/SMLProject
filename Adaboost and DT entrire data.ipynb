```python
from collections import Counter

counter = Counter(y_train)
counter
smote_oversample = SMOTE(sampling_strategy=0.5)
#random_undersample = RandomUnderSampler(sampling_strategy=0.9)
#steps = [('o', smote_oversample), ('u', random_undersample)]
#pipeline_sample = Pipeline(steps=steps)
#X3, y3 = pipeline_sample.fit_resample(X_train2,y_train2)
X3, y3 = smote_oversample.fit_resample(X_train,y_train)
counter = Counter(y3)
counter
X_train5, X_test5, y_train5, y_test5 = train_test_split(X3,y3, test_size=0.2, train_size=0.8, random_state=7)
```
