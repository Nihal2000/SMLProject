```python
# Importing the necessary packages
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE

# To ensure all outputs from a cell are shown
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

# To ensure all columns of a dataframe are shown
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("tidy_data.csv")
df
```

```python
df=df.dropna()
```

```python
df=df.drop(["id","date_of_last_rech"],axis=1)
```

```python
df
```

```python
# Install the fast_ml package for splitting the dataset
#!pip install fast_ml
```

```python
from fast_ml.model_development import train_valid_test_split

# Splitting the data into Training, validation and test sets
X_train, y_train, X_valid, y_valid, X_test, y_test = train_valid_test_split(df, target = "churn_probability", 
                                                                            train_size=0.7, valid_size=0.2, test_size=0.1, random_state=5)

# Display a few of the data sets
X_train
X_test
y_test
```

```python
# Training using the Random Forest Regressor

# Importing the model
from sklearn.ensemble import RandomForestClassifier

# Creating the model
modelRF = RandomForestClassifier(n_estimators=100, random_state=7)


# Fitting the model
modelRF.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

# Predicting Values for the training set
predsRFTrain = modelRF.predict(X_train)
print("Training Accuracy:")
accuracy_score(y_train, predsRFTrain)


# Predicting the values for validation set
predsRF = modelRF.predict(X_valid)

# Validation error
print("\nValidation Accuracy:")
accuracy_score(y_valid, predsRF)


# Predicting the values for the test set
predsRF2 = modelRF.predict(X_test)

print("\nTest Accuracy:")
accuracy_score(y_test, predsRF2)
```

```python
from sklearn.metrics import confusion_matrix

# Predicting Values for the training set
predsRFTrain = modelRF.predict(X_train)
print("Training Confusion Matrix:")
confusion_matrix(y_train, predsRFTrain)


# Predicting the values for validation set
predsRF = modelRF.predict(X_valid)

# Validation error
print("\nValidation Confusion Matrix:")
confusion_matrix(y_valid, predsRF)


# Predicting the values for the test set
predsRF2 = modelRF.predict(X_test)

print("\nTest Confusion Matrix:")
confusion_matrix(y_test, predsRF2)
```

```python
# Hyperparameter Tuning for Random Forest Regressor
from sklearn.model_selection import RandomizedSearchCV

# Hyperparameters:
n_estimators = [int(i) for i in np.linspace(start = 100, stop = 200, num = 5)]

max_features =['auto','sqrt']

max_depth = [int(i) for i in np.linspace(10,50, num = 5)]
max_depth.append(None)

min_samples_split = [2,5,10]

#min_samples_leaf = [1,2,5]

#bootstrap = [True,False]

param_grid = {"n_estimators": n_estimators, "max_features": max_features, "max_depth": max_depth, 
              "min_samples_split": min_samples_split}
```

```python
from sklearn.model_selection import train_test_split

X=df.drop(["churn_probability"],axis=1)
y=df["churn_probability"]
X_train2, X_test2, y_train2, y_test2 = train_test_split(X,y, test_size=0.2, train_size=0.8, random_state=5)
```

```python
# Defining the model
RF2 = RandomForestClassifier()

# Hyperparameter tuning using randomizedsearch
modelRF2 = RandomizedSearchCV(estimator=RF2, param_distributions=param_grid, n_iter=10, cv=3, verbose=100, random_state=7, n_jobs=-1)

# Fitting the model
modelRF2.fit(X_train2, y_train2)

# Getting the best parameters
modelRF2.best_params_
```

```python
best_RFmodel = modelRF2.best_estimator_

RFpredstrain = best_RFmodel.predict(X_train2)

print("\nTraining Confusion Matrix:")
confusion_matrix(y_train2,RFpredstrain)
print("\nTest Accuracy:")
accuracy_score(y_train2, RFpredstrain)

RFpreds = best_RFmodel.predict(X_test2)

print("\nTest Confusion Matrix:")
confusion_matrix(y_test2,RFpreds)

print("\nTest Accuracy:")
accuracy_score(y_test2, RFpreds)
```

```python
# SVM with class weights

from sklearn.svm import SVC

# define model
modelSVC = SVC(gamma='scale', class_weight='balanced')
modelSVC.fit(X_train2,y_train2)
```

```python
RFpredstrain = modelSVC.predict(X_train2)

print("\nTraining Confusion Matrix:")
confusion_matrix(y_train2,RFpredstrain)
print("\nTest Accuracy:")
accuracy_score(y_train2, RFpredstrain)

RFpreds = modelSVC.predict(X_test2)

print("\nTest Confusion Matrix:")
confusion_matrix(y_test2,RFpreds)

print("\nTest Accuracy:")
accuracy_score(y_test2, RFpreds)
```
