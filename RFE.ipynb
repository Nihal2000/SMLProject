```python
len(cols)
```

```python
data= raw_data.loc[:, ~raw_data.columns.isin(cols)]
data
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='median')
imp_mean.fit(data)
impute_data= pd.DataFrame(imp_mean.transform(data))
impute_data.columns= data.columns
```

```python
impute_data
```

```python
X= impute_data.loc[:, ~impute_data.columns.isin(['churn_probability'])]
y= impute_data.loc[:, impute_data.columns.isin(['churn_probability'])]
```

```python
X.dropna()
```

```python
rfe = RFE(estimator=DecisionTreeClassifier(max_depth= 7), n_features_to_select=1)
rfe.fit(X, y)
```

```python
rank= pd.DataFrame()

rank['column']= X.columns
rank['rfe_rank']= rfe.ranking_
```

```python
rank= rank.sort_values(by= 'rfe_rank')
```
